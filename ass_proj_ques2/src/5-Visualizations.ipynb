```python
# visualization 1: Bar plot of gender distribution
gender_counts = data['gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.savefig ('C:\\Users\\tarak\\Downloads\\PDS_ASSIGNMENT_1\\ass_proj_ques2\\results\\Gender_distribution.png')
plt.show()
```

```python
import seaborn as sns
```

```python
# visualization 2: Box plot of math scores by gender
sns.boxplot(x='gender', y='math score', data=data)
plt.title('Math Scores by Gender')
plt.savefig ('C:\\Users\\tarak\\Downloads\\PDS_ASSIGNMENT_1\\ass_proj_ques2\\results\\Math_Scores_by_Gender.png')
plt.show()
```

```python
# visualization 3: Scatter plot of math scores vs reading scores
plt.scatter(data['math score'], data['reading score'])
plt.title('Math Scores vs Reading Scores')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.savefig ('C:\\Users\\tarak\\Downloads\\PDS_ASSIGNMENT_1\\ass_proj_ques2\\results\\Math_Scores_vs_Reading_Scores.png')
plt.show()
```

```python
# visualization 4: Density plot of writing scores by lunch type
sns.kdeplot(x='writing score', data=data, hue='lunch', fill=True, alpha=.3)
plt.title('Writing Scores by Lunch Type')
plt.xlabel('Writing Score')
plt.savefig ('C:\\Users\\tarak\\Downloads\\PDS_ASSIGNMENT_1\\ass_proj_ques2\\results\\Writing_Scores_by_Lunch_Type.png')
plt.show()
```

```python
# visualization 5: Heatmap of correlation matrix
corr = data.corr(numeric_only=True)
sns.heatmap(corr, cmap='coolwarm', annot=True)
plt.title('Correlation Matrix')
plt.savefig ('C:\\Users\\tarak\\Downloads\\PDS_ASSIGNMENT_1\\ass_proj_ques2\\results\\Correlation_Matrix.png')
plt.show()
```
